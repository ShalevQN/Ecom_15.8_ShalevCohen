```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

course_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/09.04.2025/refs/heads/main/courses.csv')
students_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/09.04.2025/refs/heads/main/students.csv')
teachers_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/09.04.2025/refs/heads/main/teachers.csv')
student_courses_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/09.04.2025/refs/heads/main/teachers.csv')

for df in [course_df, students_df, teachers_df, student_courses_df]:
    df.dropna(inplace= True)
    df.drop_duplicates(inplace= True)

#merged_all = pd.concat([course_df, students_df, teachers_df, student_courses_df], axis=1, join='inner')
course_df
```

```python
# 1
gender_count = students_df['gender'].value_counts()
gender_count
```

```python
gender_count.plot(kind='bar', color=['lightblue', 'lightpink'])
plt.title('Gender Count of students')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.show()
```

```python
# 2
fig = plt.figure(figsize=(7,2)) 
ax1 = fig.add_axes([0, 1.5, 1, 1])

ax1.hist(students_df['age'], bins = 10, color = 'r')
ax1.set_title('Age disturbition in students')
ax1.set_xlabel('Age')
ax1.set_ylabel('Number of students')
```

```python
# 3
course_amount_each_category = course_df.groupby('course_category').agg(Amount_of_Courses = ('course_id', 'count'))
course_amount_each_category
```

```python
course_amount_each_category.plot(kind='bar', color=['blue'])
plt.title('Courses Amount Each Category')
plt.xlabel('Categories')
plt.ylabel('Number of courses')
plt.xticks(rotation=7)
plt.show()
```

```python
# 4
city_count = students_df.groupby('city').agg(Amount_of_students = ('city', 'count'))
print('Most amount of students: ', city_count['Amount_of_students'].idxmax())
print('Least amount of students: ', city_count['Amount_of_students'].idxmin())
```

```python
# 5
unique_names_courses = course_df.groupby('course_category').agg(Names_of_courses = ('course_name', 'unique'))
display(unique_names_courses)
```
