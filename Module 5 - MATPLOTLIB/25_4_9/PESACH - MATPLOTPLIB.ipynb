```python
import numpy as np
import matplotlib.pyplot as plt

# 1
years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

fig = plt.figure(figsize=(7,2)) 

ax1 = fig.add_axes([0, 1.5, 1, 1])
ax1.plot(height, years, color='red')
ax1.set_title('Height of a child over the years')
ax1.set_xlabel('Height')
ax1.set_ylabel('Years')
ax1.grid(True)
```

```python
# 2
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

fig = plt.figure(figsize=(7,2)) 

ax1 = fig.add_axes([0, 1.5, 1, 1])
ax1.plot(months, rain, color='blue')
ax1.set_title('Rain by Months')
ax1.set_xlabel('Months')
ax1.set_ylabel('Rain')
ax1.grid(True)

ax2 = fig.add_axes([0, 0, 1, 1])
ax2.plot(months, temperature, color='red')
ax2.set_title('Temparture by Months')
ax2.set_xlabel('Months')
ax2.set_ylabel('Temprature')
ax2.grid(True)
```

```python
# 3
grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]

fig = plt.figure(figsize=(7,2)) 
ax1 = fig.add_axes([0, 1.5, 1, 1])

ax1.hist(grades, bins = 10, color = 'b')
ax1.set_title('Grades disturbition in the test')
ax1.set_xlabel('Number of students')
ax1.set_ylabel('Grade')
```

```python
# 4
age_groups = ['10-12', '13-15', '16-18', '19-21']
students_count = [30, 45, 25, 10]

fig = plt.figure(figsize=(7,2)) 
ax1 = fig.add_axes([0, 1.5, 1, 1])

ax1.bar(age_groups, students_count, color = 'g')
ax1.set_title('Amount of students of each age group')
ax1.set_xlabel('Age Group')
ax1.set_ylabel('Number of students')
```
