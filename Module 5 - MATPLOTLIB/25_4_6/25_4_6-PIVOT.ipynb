```python
import numpy as np
from IPython.display import display
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/06.04.2025/refs/heads/main/Sales_Funnel_CRM.csv')
df
```

```python
pivot1 = pd.pivot_table(df, values = 'Licenses', index = 'Product', columns = 'Company', aggfunc = 'sum', fill_value = 0)
pivot1
```

```python
pivot2 = df.groupby('Company').agg(Total_Sale_Price = ('Sale Price', 'sum'), Licences_Sold = ('Licenses', 'sum'))
pivot2 # not really a pivot table because we dont need it, it does not have columns and index, just index.
```
