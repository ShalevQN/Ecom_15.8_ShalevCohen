```python
x_values = np.arange(0, 5, 0.1)

y_values_sin = np.sin(x_values)

y_values_cos = np.cos(x_values)

fig = plt.figure(figsize=(7,2)) 

ax1 = fig.add_axes([0, 1.5, 1, 1])
ax1.plot(x_values, y_values_sin, label='sin(x)', color='blue')
ax1.set_title('Sin Plot')
ax1.set_xlabel('x values')
ax1.set_ylabel('sin(x) values')
ax1.grid(True)
ax1.legend()

ax2 = fig.add_axes([0, 0, 1, 1])
ax2.plot(x_values, y_values_cos, label='cos(x)', color='red')
ax2.set_title('Cos Plot')
ax2.set_xlabel('x values')
ax2.set_ylabel('cos(x) values')
ax2.grid(True)
ax2.legend()
```
