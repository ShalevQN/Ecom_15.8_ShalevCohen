```python
# Seaborn
sns_df = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/exam_prep/refs/heads/main/mpg.csv")
# 1
sns_df["horsepower"] = pd.to_numeric(sns_df["horsepower"], errors="coerce")

sns.scatterplot(data= sns_df, x= "horsepower", y= "mpg", hue = "origin")
plt.show()
```

```python
# 2
filtered_df_cylinders = sns_df[sns_df["cylinders"].isin([4, 8])]
sns.barplot(data= filtered_df_cylinders, x= "cylinders", y= "mpg")
plt.show()
```

```python
sns.countplot(data= sns_df, x="origin")
plt.show()
```

```python
sns.histplot(data= sns_df, x= "weight")
plt.show()
```

```python
sns.boxplot(data= sns_df, x= "cylinders", y= "mpg")
plt.show()
```
