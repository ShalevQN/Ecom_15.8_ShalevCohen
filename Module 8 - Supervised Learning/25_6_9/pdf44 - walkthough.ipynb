```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import scipy
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ShalevQN/Ecom_15.8_ShalevCohen/refs/heads/main/Module%208%20-%20Supervised%20Learning/Advertising%20(3).csv')

df
```

```python
df['total_spent'] = df['TV'] + df['radio'] + df['newspaper']

sns.scatterplot(data= df, x= 'total_spent', y= 'sales')
```

```python
sns.regplot(data= df, x= 'total_spent', y= 'sales')
```

```python
x = df['total_spent']
y = df['sales']

betas = np.polyfit(x, y, deg=1)
betas
```

```python
expenses = 200
predicted_sales = expenses*betas[0] + betas[1] # Expenses * b1 + b0
predicted_sales
```

```python
potential_spends = np.linspace(0, 500, 100)
predicted_sales = betas[0] * potential_spends + betas[1]
plt.plot(potential_spends, predicted_sales)
plt.show()
```

```python
sns.scatterplot(data= df, x= 'total_spent', y= 'sales')
plt.plot(potential_spends, predicted_sales, color= 'red')
plt.show()
```
