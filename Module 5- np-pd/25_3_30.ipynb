```python
import numpy as np
from IPython.display import display
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/26.03.2025/refs/heads/main/movie_scores.csv')
df
```

```python
# 1
df_droppedna = df.dropna(thresh=df.shape[1] - 1, axis=0)
df_droppedna
```

```python
df_droppedna.shape[0]
```

```python
# 2
df['null_count'] = df.isna().sum(axis=1)
df
```

```python
df.sort_values('null_count', ascending = False)
```

```python
df[['first_name', 'last_name', 'null_count']].nlargest(3, 'null_count')
```

```python
# 3
nan_rows = df[df["pre_movie_score"].isna()].copy()

df["pre_movie_score"] = df["pre_movie_score"].fillna(df["pre_movie_score"].mean())
display(df)
df.loc[nan_rows.index]
```

```python
# 4
df['improvement'] = df['post_movie_score'] - df['pre_movie_score']
df
```

```python
display(df[df['improvement'] > 0])
```

```python
df.sort_values('improvement', ascending = False)
```
