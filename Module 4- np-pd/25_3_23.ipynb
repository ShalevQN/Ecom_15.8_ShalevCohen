```python
import numpy as np
import pandas as pd
from IPython.display import display

# 1
df = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/23.03.2025/main/mobile.csv")

df["First_Name"] = df["Customer Name"].apply(lambda x: str(x).split()[0])

display(df)
```

```python
# 2
carrier_mapping = {
    "050": "Pelephone",
    "051": "Golan",
    "052": "Cellcom",
    "053": "Motorola"}

df["Phone Carrier"] = df["Phone Number"].astype(str).str[:3].replace(carrier_mapping)

print(df[["Phone Number", "Phone Carrier"]])
```

```python
# 3
print("Costumers per carrier:\n", df["Phone Carrier"].value_counts())

print("\nUnique carriers:\n", df["Phone Carrier"].unique())

print("\nNumber of unique carriers:", df["Phone Carrier"].nunique())
```

```python
df["Type Of Customer"] = df.apply(lambda row: 'talker' if row['Talked'] > row['Internet'] else 'surfer', axis=1)
print(df[["Customer Name", "Talked", "Internet", "Type Of Customer"]])
```

```python
# 5

print("A\n", df.sort_values("Talked"))
```

```python
print("B\n", df.sort_values(by = ["Talked", "Internet"]))
```

```python
# 6
print(df.nlargest(5, "Talked"))

print(df.nsmallest(3, "Internet"))
```

```python
# 7
duplicates = df[df.duplicated()]
print(duplicates[["Phone Number", "Customer Name"]])

print(df["Customer Name"].drop_duplicates)
```

```python
winners = df.sample(n=3)

print(winners[["Customer Name", "Phone Number"]])
```
