```python
import numpy as np
from IPython.display import display
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/02.04.2025/refs/heads/main/sport_athletes.csv')
df
```

```python
# 1
# a 
df.rename(columns = {'Weight(kg)': 'max_weight'}, inplace=True)
# b
df.rename(columns = {'Height(cm)' : 'max_height'}, inplace=True)
# c
df.rename(columns = {'Wins' :'max_wins'}, inplace=True)

print('max height:\n')
print(df.groupby('Sport Type')['max_height'].nlargest(1), '\n')

print('lowest weight height:\n')
print(df.groupby('Sport Type')['max_weight'].nsmallest(1), '\n')

print('highest wins:\n')
print(df.groupby('Sport Type')['max_wins'].nlargest(1), '\n')
```

```python
# 2
df.groupby('Sport Type')['Height(cm)'].nlargest(2)
```

```python
# 3
# a
groupped = df.groupby(['Sport Type', 'Country']).agg({'Wins': 'max', 'Height(cm)': 'max', 'Weight(kg)': 'max'})
groupped
```

```python
# a - alternative
df.groupby(['Sport Type', 'Country']).agg({'Wins': 'max'})
```

```python
# b
df.loc[(df['Country'] == 'USA') & (df['Sport Type'] == 'Basketball')].max()
```

```python
# c
groupped.xs('Soccer', level = 'Sport Type')
#df.loc[df['Sport Type'] == 'Soccer'].max()
```

```python
# d
df[df['Sport Type'].isin(['Tennis', 'Soccer'])].max()
```

```python
# e
df.swaplevel(i= 'Sport Type', j = 'Country', axis = 1)
```

```python
# f
df.agg({
    'Wins': ['min', 'max', 'mean', 'std'],  
    'Height(cm)': ['min', 'max', 'mean', 'std']})
```

```python
# 4
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })

pd.concat([df, new_rows], axis=0, ignore_index=False)
```
