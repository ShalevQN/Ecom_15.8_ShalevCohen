```python
import numpy as np
from IPython.display import display

# 1
np.random.seed(80)

nums1 = np.random.randint(1, 101, (4, 4))
print(nums1)
# a
print("Bigger than 50:", nums1[(50 < nums1)])

# b
#print("Sum of collums:", sum(nums1[:, x] for x in range(len(nums1[0]))))
print("Sum of collums:", nums1.sum(axis = 1))

#print("Sum of rows:", sum(nums1[x] for x in range(len(nums1[:, 0]))))
print("Sum of rows:", nums1.sum(axis = 0))

# c
# i
print("Max Rows:", np.max(nums1, axis = 1))
print("Max Collums:", np.max(nums1, axis = 0))
# ii
print("Min Rows:", np.min(nums1, axis = 1))
print("Min Collums:", np.min(nums1, axis = 0))
# iii
print("Mean Rows:", np.mean(nums1, axis = 1))
print("Mean Collums:", np.mean(nums1, axis = 0))
```

```python
import pandas as pd

# 2
new_games = {
"Final Fantasy VII Rebirth": 69.99,
"Astro Bot": 59.99,
"Star Wars Outlaws": 69.99,
"Silent Hill 2 Remake": 69.99,
"Elden Ring: Shadow of the Erdtree": 39.99,
"Dragon Age: The Veilguard": 69.99,
"Metaphor: ReFantazio": 59.99
}

new_games_pd = pd.Series(new_games)
display(new_games_pd)

# a
print("Metaphor: ReFantazio's price: ", new_games_pd["Metaphor: ReFantazio"])

# b
print(new_games_pd.keys())

# c
print(f"Item in index 3 is {new_games_pd.keys()[3]}, its price is: ", new_games_pd[new_games_pd.keys()[3]])

# d
print("All new games prices combined: ", new_games_pd.sum())
```
